```python
plt.figure(figure=(12,6))
sns.barplot(x='Sex',y='Age',data=data)
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Passengers')
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figure=(20,10))
sns.barplot(x='Age',y='SibSp',data=data)
plt.xlabel('Age')
plt.ylabel('SibSp')
plt.title('Passengers')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
data.Sex
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', data=data)
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the distribution of fares
plt.figure(figsize=(10, 6))
sns.histplot(data['Fare'], bins=30, kde=True)
plt.title('Distribution of Fares')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Pclass'], bins=3, kde=True)
plt.title('Distribution of Passenger Class (Pclass)')
plt.xlabel('Pclass')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['SibSp'], bins=8, kde=True)
plt.title('Distribution of Number of Siblings/Spouses aboard (SibSp)')
plt.xlabel('SibSp')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Parch'], bins=6, kde=True)
plt.title('Distribution of Number of Parents/Children aboard (Parch)')
plt.xlabel('Parch')
plt.ylabel('Frequency')
plt.show()
```
